```python
# 1. Class to generate data - data generator - numerical experiment - synthetic
# 2. Class to store data (synthetic or real)
# 3. Class to model data (mathematical relations)

# 1. Data generator recipe: a) physical model, b) experiment setup and process
# What happens during experiment? Entropy is increasing

# Sprint hanging vertically, mass is attached, we measure elongation of spring.
# date time to log time

class Experiment:

    def __init__(self):
        self.mass = [0, 1, 2, 3, 4]                   # [kg]
        self.length = [0.1, 0.15, 0.18, 0.24, 0.3]    # [m]
        assert len(self.mass) == len(self.length)

    def run(self, file_name):
        fout = open(file_name, 'w')
        fout.write('# mass length\n')
        fout.write('# [kg] [m]\n')
        # zip creates a list of tuples
        # mass, length is a tuple
        for mass, length in zip(self.mass, self.length):
            fout.write(f'{mass},{length}\n')
        fout.close()


experiment = Experiment()
experiment.run(file_name='exp1.txt')
```

```python
import matplotlib.pyplot as plt

class Data:

    def __init__(self):
        self.mass = []
        self.length = []
        self._use_by_me_only = None              # e.g. agreed that if method (function) or attribute (variable) starts with underscore, it will not be used outsise of the class
        self._dont_use_this_externally = None

    def load(self, file_name):
        with open(file_name, 'r') as fin:
            for line in fin:
                if line[0] != '#':
                    mass, length = line.split(',')
                    self.mass.append(float(mass))
                    self.length.append(float(length))

    def plot(self, fig_name):
        plt.figure(figsize=(3*1.6, 3))  # in inches
        # TODO subplots adjust
        plt.scatter(self.mass, self.length, c=['r', 'g', 'b', 'c', 'm'], s=20 * 50)
        plt.xlabel('$mass$ [kg]')
        plt.ylabel('$length$ [m]')
        # TODO add the tick marks inside
        # TODO tick marks should be on all sides top, bottom, left, right
        # TODO tick-marks 3-5 per axis
        plt.savefig(fig_name, dpi=300)


data = Data()
data.load(file_name='exp1.txt')
data.plot(fig_name='exp1.png')
data.mass
```

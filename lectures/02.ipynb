```python
# create class object that contains variables - attributes
# lets add method to print those attributes

# within class self.attribute and function(self, )

class Person:
    name = 'UNDEFINED'
    surname = 'UNDEFINED'

    def show(self):
        print('Name:', self.name)
        print('Surname:', self.surname)


person1 = Person()
person2 = Person()

person1.name = 'Petras'
person1.surname = 'Petraitis'

# print(person1)

person1.show()
```

```python
# when object is created from template e.g. MyClass,
# the brackets () after the template name MyClass()
# do call the __init__(self) method of class.


# create a new class template

class Class5:

    def __init__(self):
        print('Our customised init has run')

    def show(self):
        print('all good')


obj5 = Class5()

obj5.show()   # call our "show" method

# attributes are properties of person
# methods are actions person can do
```

```python
class Ball:

    def __init__(self, input_name, input_mass, input_size):
        self.name = input_name
        self.mass = input_mass
        self.size = input_size

    def __repr__(self):
        name_str = 'name: ' + str(self.name)
        mass_str = 'mass: ' + str(self.mass)
        size_str = 'size: ' + str(self.size)
        something = name_str + ', ' + mass_str + ', ' + size_str
        return something


ball = Ball(
    input_name='vardas',
    input_mass=1,
    input_size=2
)

ball
```

```python
class Dataset:

    def __init__(self):
        self.x_list = [1, 2, 3, 4, 5]   # mass of the object
        self.y_list = [0.10, 0.21, 0.28, 0.42, 0.55]
        assert len(self.x_list) == len(self.y_list)
        # TODO check that type of variables is numerical

    def save(self, filename):
        fout = open(filename, 'w')
        for x, y in zip(self.x_list, self.y_list):  # zip lists
            fout.write(f'{x} {y}\n')   # f-string


dataset = Dataset()
dataset.save(filename='test.txt')

# add x and y lists to store measurements
```

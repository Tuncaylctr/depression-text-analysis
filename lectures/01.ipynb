```python
# image size

# bits 0/1
# byte 8 bits: 0000,0000 -> 8 memory cells for one byte
# 0000,0001 -> 1
# 0000,0010 -> 2
# 0000,0011 -> 3

# 7654,3210 position encodes powers of 2 and we add them up

# 1000,0000 -> 2^7 -> 128
# 1111,1111 ->
# 2^7 + 2^6 + 2^5 + ... + 2^1 = 255
# 2^8 = 256

# 1 byte encodes numbers between 0 and 255. There are 256 values

# 2 bytes |0000,0000|0000,0000| -> 16
# 2^16 = 65536
# max 65536 - 1

# 1024 bytes is KB
# 1024 kbytes is MB
```
